```python
# =========================================
# 1. IMPORT LIBRARIES
# =========================================
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# =========================================
# 2. INPUT DATA (JOB DESCRIPTION + RESUMES)
# =========================================
job_description = """
Looking for a Python developer with experience in machine learning,
data analysis, NLP, and knowledge of TensorFlow or Scikit-learn.
"""

resumes = [
    "Experienced Python developer with machine learning and data analysis using sklearn and pandas.",
    "Frontend developer skilled in HTML CSS JavaScript and React.",
    "Data scientist with Python TensorFlow deep learning and AI experience.",
    "Software engineer with Java backend development and SQL.",
    "Machine learning engineer with NLP experience and strong Python skills."
]

# =========================================
# 3. COMBINE TEXT DATA
# =========================================
documents = [job_description] + resumes

# =========================================
# 4. TF-IDF VECTORIZATION
# =========================================
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(documents)

# =========================================
# 5. COSINE SIMILARITY
# =========================================
similarity_scores = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])

# =========================================
# 6. RANK RESUMES
# =========================================
scores = similarity_scores.flatten()

ranked_resumes = sorted(
    list(enumerate(scores)),
    key=lambda x: x[1],
    reverse=True
)

print("📊 Resume Ranking:\n")

for rank, (index, score) in enumerate(ranked_resumes, start=1):
    print(f"Rank {rank}: Resume {index+1} | Score: {score:.4f}")

# =========================================
# 7. BEST CANDIDATE
# =========================================
best_index = ranked_resumes[0][0]

print("\n🏆 Best Resume:\n")
print(resumes[best_index])

# =========================================
# 8. OPTIONAL: CHECK CUSTOM RESUME
# =========================================
def check_resume(resume_text):
    new_docs = [job_description, resume_text]
    new_tfidf = vectorizer.transform(new_docs)
    score = cosine_similarity(new_tfidf[0:1], new_tfidf[1:2])[0][0]
    print("\n📌 Custom Resume Score:", round(score, 4))

# Test with your own resume
check_resume("Python developer with NLP and machine learning experience")
```
